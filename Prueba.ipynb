### Plots

```julia
x = range(-2*pi, stop = 2*pi, length = 50)
y1 = sin.(x)
y2 = cos.(x)

plt_sincos = plot(x,[y1 y2], xlabel="x", ylabel="f(x)", title = "Función Seno y Coseno", 
                  color=["blue" "red"], label=["sin(x)" "cos(x)"], legend = true, 
                  linewidth = 2, grid = true)
display(plt_sincos)

#Las condiciones que le ponemos al gráfico se parecen bastante a la graficación en RStudio, se ve chévere y se logra cachar
#que es lo que está haciendo en el gráfico
```
